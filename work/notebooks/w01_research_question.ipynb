## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Question: What performance archetypes exist across the content inventory,
and what action fits each archetype?**

**Decision this improves: Which group of pages a content team should focus
on first, and what general action (protect, improve, monitor, prune)
fits that group.**

**Unit of analysis: one content page (one row per content_id).**

**Action someone could take: A content strategist could use the cluster
labels to triage hundreds of pages quickly — e.g. "these 40 pages are in
the stale-but-visible cluster, review them for refresh" — instead of
reviewing every page individually.**

**Cost of a wrong call: If a page is placed in the wrong cluster, the team
might spend review time on a low-value page (wasted effort) or miss a
page that actually needed attention (opportunity cost). Since this is a
triage tool, not an individual verdict, the cost of one wrong cluster
assignment is moderate, not severe — but a systematically wrong cluster
definition could misdirect the whole team's effort.**

**Why data/ML helps: With 30,000+ pages, no one can manually inspect each
one. Clustering surfaces natural groupings from the numbers themselves,
rather than relying on one hand-picked rule for every page**

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", df.shape[0])
print("Columns:", df.shape[1])

# Number 1: how varied is engagement across pages?
print("\nEngagement rate spread:")
print(df["engagement_rate"].describe()[["mean", "std", "min", "max"]])

# Number 2: how many pages are stale but still visible? (a likely archetype)
stale_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()
print(f"\nStale-but-visible pages: {stale_visible} ({stale_visible/len(df)*100:.1f}% of all pages)")

# Number 3: trend_direction breakdown (shows there are already distinct behavior groups)
print("\nTrend direction breakdown:")
print(df["trend_direction"].value_counts())
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

**What I CAN claim:**
- Observed groupings in the data based on measurable signals (impressions,
  CTR, freshness, engagement, etc.)
- Directional patterns — e.g. "this cluster tends to have low engagement
  and high impressions"
- A decision-support tool that helps a reviewer prioritize, not a
  guaranteed verdict

**What I CANNOT claim:**
- That these clusters represent true "meaning" of the content (this is
  metric-based clustering, not semantic/text-based clustering)
- That any cluster proves a Google ranking factor
- That moving a page out of a "declining" cluster will cause it to recover
  — that would require a causal experiment, not just clustering
- That the clusters are fixed truth — they are a lens, and may shift if
  I re-run with different features or a different k
